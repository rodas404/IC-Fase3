Imports

```python
import numpy as np
from sklearn.utils import shuffle
from SwarmPackagePy import pso, fwa, animation, animation3D
from tensorflow.keras import models, layers, optimizers, callbacks
from tensorflow.keras.applications import VGG16, Xception
```

Carregar datasets e baralha-los

```python
train_images = np.load('arraysTrainTest/train_images71.npy')
train_labels = np.load('arraysTrainTest/train_labels71.npy')
train_images, train_labels = shuffle(train_images, train_labels, random_state=25)

test_images = np.load('arraysTrainTest/test_images71.npy')
test_labels = np.load('arraysTrainTest/test_labels71.npy')
```

Rede neuronal a otimizar

```python
def fitness_function(params):
    neurons, learn_rate= params
    
    base_model = Xception(weights='imagenet', include_top=False, input_shape=(71, 71, 3))

    callback = callbacks.EarlyStopping(monitor='loss', patience=3)
    
    for layer in base_model.layers:
        layer.trainable = False
    
    model = models.Sequential([
        base_model,
        layers.Flatten(),
        layers.Dense(neurons, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(neurons, activation='relu'),
        layers.Dense(5, activation='softmax')
    ])

    
    model.compile(optimizer=optimizers.Adam(learning_rate = learn_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    
    model.fit(train_images, train_labels, epochs=10, validation_split=0.2, batch_size=64)
    error, accuracy = model.evaluate(test_images, test_labels)
    
    return 1-accuracy
```

Chamar pso

```python
lb = [64, 0.0001] #n_neuronios e learning_rate
ub = [256, 0.01]
alh = pso(3, fitness_function, lb, ub, 2, 10, w=0.2, c1=0.9, c2=0.9)
                            
best_params = alh.get_Gbest()
print("Melhores parametros:", best_params)
best_cost = -fitness_function(best_params) + 1
print("Com accuracy de:", best_cost)
```
